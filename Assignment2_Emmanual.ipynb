```python
#6.Detection Of Outlier-BoxPlot
import pandas
import numpy
Churn = pandas.read_csv("Churn_Modelling.csv")
numeric_col = ['Balance','EstimatedSalary']
categorical_col = ['Age','Tenure','NumOfProducts','HasCrCard','IsActiveMember']
Churn.boxplot(numeric_col)
```

```python
#Removal of Outliers
for x in ['Balance']:
    q75,q25 = np.percentile(Churn.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    Churn.loc[Churn[x] < min,x] = np.nan
    Churn.loc[Churn[x] > max,x] = np.nan
Churn.isnull().sum()
```

```python
Churn = Churn.dropna(axis = 1)
Churn.isnull().sum()
```
